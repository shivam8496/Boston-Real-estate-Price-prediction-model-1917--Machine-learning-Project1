```python
## train test function
```

```python
import numpy as np
def train_test_split(data,test_ratio):
    np.random.seed(42) ## saves the state of randomness
    shuffled =np.random.permutation(len(data))
    test_set_size =int(len(data) * test_ratio)
    test_data=shuffled[:test_set_size]
    traing_data =shuffled[test_set_size:]
    return data.iloc[traing_data] ,data.iloc[test_data]
```

```python
train_set,test_set = train_test_split(housing,0.2)
print("train len =>",len(train_set),"  Test len ==>",len(test_set))
```
